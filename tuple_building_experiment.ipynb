```python
from rtgenerator.tuple_generator import generate_random_state_size
from rtgenerator.tree_generator import generate_random_tree_height
import numpy as np
import sklearn.linear_model
import math
import scipy.optimize
%pylab inline
```

```python
def run_experiment(func, children_num_generator, step_nums, iterations_num=10000):
    print 'iterations=', iterations_num, 'steps num=', step_nums
    results = []
    for n in step_nums:
        print 'step_num=', n
        samples = [func(children_num_generator, n) for i in xrange(iterations_num)]
        results.append(samples)
    return results
```

```python
import pickle
def save_to(filename, step_nums, results):
    with open(filename + '_steps.pickle', "w") as f_pickle:
        pickle.dump(step_nums, f_pickle)
    with open(filename + '_res.pickle', "w") as f_pickle:
        pickle.dump(results, f_pickle)
    with open(filename + '.txt', "w") as f:
        f.write("Generator: 1\n")
        f.write("Steps: %s\n" % step_nums.__str__())
        for n, r in enumerate(results):
            f.write("for step %d results: %s\n" %(step_nums[n], r))
            
def load_from(filename):
    with open(filename + '_steps.pickle', "r") as f_pickle:
        step_nums = pickle.load(f_pickle)
    with open(filename + '_res.pickle', "r") as f_pickle:
        results = pickle.load(f_pickle)
    return step_nums, results  
```

```python
# plt.semilog(step_nums, np.mean(results, axis=1))
def plt_results(step_nums, results, filename=None):
    plt.figure(figsize=(10, 7))
    plt.semilogx(step_nums, np.mean(results, axis=1), label='mean')
    plt.semilogx(step_nums, np.median(results, axis=1), label='median')
    plt.semilogx(step_nums, np.min(results, axis=1), label='min')
    plt.semilogx(step_nums, np.max(results, axis=1), label='max')
    plt.semilogx(step_nums, np.percentile(results, 99.9, axis=1), label='quantile 0.999')
    plt.legend(loc='best')
    plt.xlabel('Steps num')
    plt.ylabel('Tree height')
    if filename is not None:
         plt.savefig(filename + '.jpg')
#     plt.title('Log to log plot of tree height with respect to the number of steps taken')
```

```python
step_nums = [2 ** x for x in xrange(16)]
filename = "gen_1_range_16_10000"
results = run_experiment(generate_random_state_size, lambda: 1, step_nums, 10000)
save_to(filename, step_nums, results)
plt_results(step_nums, results, filename)
```

```python
def log_x_slopes(step_nums, results):
    log_steps = np.log2(step_nums)
    log_step_inc = log_steps[1:] - log_steps[:-1]
    tasks = []
    tasks.append(('max', np.max))
    tasks.append(('quantile 0.999', lambda x, axis: np.percentile(x, 99.9, axis)))
    tasks.append(('mean', np.mean))
    tasks.append(('median', np.median))
    tasks.append(('min', np.min))
    out = {}
    for name, func in tasks:
        f_results = func(results, axis=1)
        slopes = (f_results[1:] - f_results[:-1]) / log_step_inc
        out[name] = np.mean(slopes[8:])
    return out
log_x_slopes(step_nums, results)
```

```python
def theoretical_slope(expectation):
    alpha = expectation + 1
    eq_func = lambda x: x*math.exp(x) - 1.0 / expectation
    gamma = scipy.optimize.brentq(eq_func, 0, 10)
    return alpha, gamma, 1.0 / (alpha * gamma)
theoretical_slope(10000)
```
