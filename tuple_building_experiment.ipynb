```python
import src.generator as generator
import numpy as np
%pylab inline
```

```python
def height(a):
    if a[-1] != 0:
        return len(a)
    return np.argmin(a)
```

```python

step_nums = [2 ** x for x in xrange(10)]
results = []
for n in step_nums:
    samples = [generator.generate_random_state(lambda: 2, generator.weighted_level_generator, num_steps=n) for i in xrange(1000)]
    results.append([height(s) for s in samples])
```

```python
m = np.mean(results, axis=1)
```
